## 2.Предобработка

```python
# Сортировка значения датафрейма по дате
df.sort_values(by=['id','date'], inplace=True)
```

```python
# Процент пропусков в данных
missing_percent = round(100 * df.isnull().sum() / len(df),2)
display(missing_percent[missing_percent > 0].sort_values(ascending=False))
```

Пропуски в столбцах можно отнести к разным категориям.

**MNAR**

Пропуски в числовом столбце `monthly_income_amt` и `children_cnt`. Пользователи из личных соображений не указывают уровень дохода. Заполним значением заглушкой -999.

В категориальных столбцах `education_level_cd и marital_status_cd` уровня образования и семейного положения стоит заполнить категорией `No_info`

**MCAR**

В столбцах пользовательсого поведения (`posts,reacts,comments, reads`) отстутствие значений может говорить об отсутствии активности. Эти значения можно заполнить 0.

Пропуски в количественных столбцах `lon_balance_amt` и `ccr_balance_amt` может говорить об отсутствии взаимодейстий с заемными средствами и кредитными счетами. Заполним 0.

Пропуски в столбце `cor_balance_amt` не станем удалять, чтобы сохранить данные закаждый день для каждого пользователя. Заполним медианным значением.

```python
# Заполнение пропусков:
  # Категориальных данные
df['education_level_cd'] = df['education_level_cd'].fillna('No_info')
df['marital_status_cd'] = df['marital_status_cd'].fillna('No_info')
  # Количественные данные  по балансу заполним медианными значениями и по общественным характеристикам заполним нулем
quantity_cols_0 = ['lon_balance_amt', 'ccr_balance_amt']
quantity_cols_median = ['cor_balance_amt']
for col in quantity_cols_median:
  df[col] = df[col].fillna(df[col].notnull().median())
for col in quantity_cols_0:
  df[col] = df[col].fillna(0)
social_cols = ['posts', 'comments', 'reacts', 'reads']
for col in social_cols:
  df[col] = df[col].fillna(0)

# Пропуски MNAR
df['monthly_income_amt'] = df['monthly_income_amt'].fillna(-999)
df['children_cnt'] = df['children_cnt'].fillna(-999)
```

```python
# Преобразование df к правильному типу.
# Создаем копию DataFrame для безопасного преобразования
df_converted = df.copy()
# Функция для безопасного преобразования в числовой формат
def safe_numeric_conversion(series, is_integer=False):
    # Сначала преобразуем в строку и очистим
    cleaned = series.astype(str).str.replace(' ', '').str.replace(',', '.')
    # Преобразуем в числовой формат
    numeric_series = pd.to_numeric(cleaned, errors='coerce')
    # Для целых чисел преобразуем к целому типу
    if is_integer:
        # Сначала проверим, есть ли нецелые значения
        non_integer_mask = (numeric_series % 1 != 0) & (~numeric_series.isna())
        if non_integer_mask.any():
            print(f'Предупреждение: в колонке найдены нецелые значения: {series.name}')
            # Округляем до целых
            numeric_series = numeric_series.round()
        # Преобразуем в целочисленный тип с поддержкой NaN
        numeric_series = numeric_series.astype('Int64')
    return numeric_series
# Преобразование колонок с суффиксом '_amt' в float
amt_cols = [col for col in df.columns if col.endswith('_amt')]
for col in amt_cols:
    df_converted[col] = safe_numeric_conversion(df[col])
# Преобразование колонок с суффиксом '_cnt' в int
cnt_cols = [col for col in df.columns if col.endswith('_cnt')]
for col in cnt_cols:
    df_converted[col] = safe_numeric_conversion(df[col], is_integer=True)
# Преобразование флагов в категориальные значения
category_cols = [col for col in df.columns if col.endswith('_cd')]
for col in category_cols:
    df_converted[col] = df_converted[col].astype('category')
# Преобразование флагов в булевые значения
boolean_cols = [col for col in df.columns if col.endswith('_flg')]
for col in boolean_cols:
    df_converted[col] = df_converted[col].astype(bool)
# Преобразование социальных метрик в int
social_cols = ['posts', 'comments', 'reacts', 'reads']
for col in social_cols:
    df_converted[col] = safe_numeric_conversion(df[col], is_integer=True)
# Проверка преобразования
display(df_converted.info())
```

```python
# Проверяем на дубликаты
df_converted.duplicated(subset = ['id','date']).sum()
```

```python
# Сравним размер до и после оптимизации
def get_memory_usage(df):
    return df.memory_usage(deep=True).sum() / 1024**2  # в МБ
get_memory_usage(df_converted)
display(f"Размер DataFrame до оптимизации: {get_memory_usage(df_converted):.2f} МБ")

# Преобразование счетчиков в более компактные типы
int_columns = [
    'children_cnt', 'trade_order_rus_sec_cnt', 'trade_order_rus_bon_cnt',
    'trade_order_forts_cnt', 'trade_order_cur_cnt', 'trade_order_etf_cnt',
    'trade_order_fnd_cnt', 'trade_order_opt_cnt', 'trade_order_ore_prc_cnt',
    'trade_order_ore_sel_cnt', 'trade_order_tracking_cnt', 'posts',
    'comments', 'reacts', 'reads'
]
for col in int_columns:
    # Определяем подходящий тип на основе диапазона значений
    max_val = df_converted[col].max()
    if max_val < 128:
        df_converted[col] = df_converted[col].astype('int8')
    elif max_val < 32768:
        df_converted[col] = df_converted[col].astype('int16')
    else:
        df_converted[col] = df_converted[col].astype('int32')
# Преобразование финансовых показателей в float32
float_columns = [
    'monthly_income_amt', 'in_payment_rub_amt', 'out_payment_rub_amt',
    'uncovered_position_amt', 'turnover_rus_bon_amt', 'turnover_rus_sec_amt',
    'turnover_forts_amt', 'turnover_cur_amt', 'turnover_etf_amt',
    'turnover_fnd_amt', 'turnover_opt_amt', 'turnover_ore_prc_amt',
    'turnover_ore_sel_amt', 'turnover_tracking_amt', 'portf_total_amt',
    'portf_mex_amt', 'portf_rus_sec_amt', 'portf_rus_bon_amt',
    'initial_margin_amt', 'portf_cur_amt', 'portf_etf_amt', 'portf_fnd_amt',
    'portf_opt_amt', 'portf_ore_amt', 'ccr_balance_amt', 'cor_balance_amt',
    'lon_balance_amt'
]

for col in float_columns:
    df_converted[col] = df_converted[col].astype('float32')

display(f"Размер DataFrame после оптимизации: {get_memory_usage(df_converted):.2f} МБ")
```
